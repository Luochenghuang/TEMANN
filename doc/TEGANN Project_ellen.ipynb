There are duplicate formula names as data was gathered at different temperatures.

```python
def remove_duplicates(values):
    """Clean duplicate formula names out of list"""
    output = []
    seen = set()
    for value in values:
        # If value has not been encountered yet,
        # ... add it to both list and set.
        if value not in seen:
            output.append(value)
            seen.add(value)
    return output

# Remove duplicates from this list.
values = formula_list
dups_removed_formulas = remove_duplicates(values)
print(dups_removed_formulas[:10])
```

```python
def find_bandgap_e(series):
    data = []
    for idx in series:
        formula = idx
        data = a.query(criteria={"pretty_formula": formula}, properties=["band_gap"])
    data_df = pd.DataFrame(data)
    return data_df
```
